# Modelo Não-Supervisionado

Primeiro vamos importar todas ferramentas que vamos utilizar

```
%load_ext google.colab.data_table
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

Agora vamos abrir um *dataset*

Um *dataset* é um conjunto de dados

Esse *dataset* é sobre espécies de plantas com os tamanhos de suas sépalas e pétulas

```
iris_df = datasets.load_iris()
df = pd.DataFrame(iris_df.data, columns=iris_df.feature_names)
df
```

Como esse é um exemplo, ele já vem com as respostas de quais plantas são de cada espécie, abaixo agrupadas por cores

```
x_axis = iris_df.data[:, 0]  # Sepal Length
y_axis = iris_df.data[:, 1]  # Sepal Width


plt.scatter(x_axis, y_axis, c=iris_df.target)
plt.show()
```

Agora vamos tentar descobrir a espécie usando os KMeans, que separa as plantas com características mais parecidas

```
model = KMeans(n_clusters=3)
model.fit(iris_df.data)
all_predictions = model.predict(iris_df.data)

plt.scatter(x_axis, y_axis, c=all_predictions)
plt.show()
```
